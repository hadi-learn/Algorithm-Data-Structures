**Reverse Array**

```python
def reverse_list_1(lst):
  return lst[::-1]

def reverse_list_2(lst):
  n = len(lst) - 1
  reversed_list = []
  while n >= 0:
    reversed_list.append(lst[n])
    n -= 1
  return reversed_list
```

1.   Check the frequency of each letter on each string
2.   Anagram is true if the sorted string is equal for both string



```python
def anagram_1(string1, string2):
  special_chars = [" ", ".",",", "?", "!","-", "_", ":","&", ";",]
  string1 = string1.replace(" ","").replace(",", "").replace(".", "").lower()
  for char in special_chars:
    string2 = string2.replace(char, "").lower()
  
  print(string1)
  print(string2)

  return sorted(string1) == sorted(string2)
```

```python
def pair_sum(lst, target):
  index = 0
  n = len(lst)
  pair = set()
  while index < n:
    index_2 = index + 1
    while index_2 < n:
      if lst[index] + lst[index_2] == target:
        if lst[index] not in pair:
          print(f"added: {lst[index]} and {lst[index_2]}")
          # pair.add((lst[index], lst[index_2]))
          ## the above tuple add cannot be implemented since the tuple (1, 3) and (3, 1)
          ## will be counted as different value in a set
          ## while using the min and max function will always put the least integer as the first value of the tuple,
          ## so the tuple (1, 3) and (3, 1) will be regarded the same
          pair.add( ((min(lst[index], lst[index_2])), (max(lst[index], lst[index_2]))) )
          print(pair)
        
      index_2 += 1
    index += 1

  print("returned value:")
  # return len(pair)
  # return pair
  return print("\n".join(map(str, list(pair))))

## the above solution is O(n^2) so it's not a good solution
```

```python
def largest_continuous_sum(lst):
  if len(lst) == 0:
    return 0
  largest = continuous_sum = lst[0]
  first_sum = lst[0] + lst[1]
  for number in lst[1:]:
    continuous_sum += number
    if continuous_sum > largest:
      largest = continuous_sum
  
  last_sum = continuous_sum

  print(f"first_sum: {first_sum}")
  print(f"last_sum: {last_sum}")

  return largest

## this solution only for the default index from the given array
## this solution is Big O(n)
```

Another approach is to use the max function to compare between the current sum value with the largest sum value

```python
def largest_continuous_sum_3(lst):
  if len(lst) == 0:
    return 0
  
  current_sum = largest_sum = lst[0]
  largest_sum_segment_tracker = []
  for number in lst[1:]:
    next_sum = current_sum + number

    if largest_sum > next_sum:
      largest_sum_segment_tracker.append(largest_sum)

    current_sum = max((current_sum + number), number) # compare the value
    
    largest_sum = max(largest_sum, current_sum) # compare the value

  largest_sum_segment_tracker.append(current_sum)
  print(f"first sum segment: {largest_sum_segment_tracker[0]}")
  print(f"last sum segment: {largest_sum_segment_tracker[-1]}")
  print(f"sum segment: {largest_sum_segment_tracker}")
  
  return print(f"largest_sum_segment: {largest_sum}")
```

**Sentence Reversal**

```python
def my_sentence_reversal(sentence):
  splitted = sentence.split(" ")
  remove_space = []
  reverse_sentence = []
  for word in splitted:
    if len(word) == 0:
      continue
    else:
      remove_space.append(word)

  index = len(remove_space) - 1
  while index >= 0:
    reverse_sentence.append(remove_space[index])
    index -= 1


  print(sentence)
  print(" ".join(remove_space[::-1]))
  print(" ".join(reverse_sentence))
```

```python
my_sentence_reversal("let's   check this out.")
```

```python
def sentence_reversal(sentence):

  length = len(sentence)
  words = []
  char_to_ignore = [" "]

  i = 0

  while i < length:

    if sentence[i] not in char_to_ignore:

      word_start = i

      while i < length and sentence[i] not in char_to_ignore:

        i += 1
      
      words.append(sentence[word_start:i])
  
    i += 1

  print(words) # to check the list words

  ## code to reverse the words
  words_reversed = []

  index = len(words) - 1

  while index >= 0:
    words_reversed.append(words[index])
    index -= 1

  return print(" ".join(words_reversed))
```

```python
def my_string_compression(string1):
  container = {}
  print(string1)
  for char in string1:
    try:
      container[char] += 1
    except KeyError:
      container[char] = 1

  compress = []
  key = list(container.keys())
  value = list(container.values())
  for i in range(len(container)):
    compress.append(key[i])
    compress.append(str(value[i]))

  print(compress)
  print("".join(compress))
```

```python
def compress(s):
  run = ""
  length = len(s)
  
  if length == 0:
    return ""
  
  if length == 1:
    return s + "1"
  
  counter = 1
  index = 1

  while index < length:
    if s[index] == s[index - 1]:
      counter += 1
    else:
      run = run + s[index - 1] + str(counter)    # build the pair of character and counter
      counter = 1    # reset the counter if the character different
    
    index += 1

  run = run + s[index-1] + str(counter)

  return run
```

**Unique Character In a String**

```python
def my_unique(string1):
  unique = {}
  for char in string1:
    try:
      unique[char] += 1
      return False
    except KeyError:
      unique[char] = 1
      
  return True
```
